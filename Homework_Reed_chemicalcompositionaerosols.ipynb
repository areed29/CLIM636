### Look at the full time period of the data record, and analyze the long-term trend of PM2.5, and its five most major components at GUMO1.

```python
f='/home/areed29/CLIM636/IMPROVE_allyears.csv'
df=pd.read_csv(f)
#print(df.to_string())
```

```python
columns = ['ALf_Val', 'ammNO3f_Val', 'ammSO4f_Val', 'ASf_Val', 'BRf_Val', 'CAf_Val',
           'EC1f_Val', 'EC2f_Val', 'EC3f_Val', 'ECf_Val', 'OC1f_Val', 'OC2f_Val',
           'OC3f_Val', 'OC4f_Val', 'OMCf_Val', 'OCf_Val', 'CHLf_Val', 'CLf_Val',
           'CRf_Val', 'CUf_Val', 'FEf_Val', 'PBf_Val', 'MGf_Val', 'MNf_Val',
           'NIf_Val', 'NO3f_Val', 'N2f_Val', 'Kf_Val', 'RBf_Val', 'SEf_Val', 
           'SIf_Val', 'NAf_Val', 'SOILf_Val', 'SRf_Val', 'SO4f_Val', 'Sf_Val', 
           'TIf_Val', 'Vf_Val', 'ZNf_Val', 'ZRf_Val']
```

```python
x=np.arange(0,4157,1)
print(np.shape(x))
```

```python
#m and b should be of length 40
```

```python
data_kept_all=[]
all_m=[]
all_b=[]
all_yfit=[]
for i in columns:
    print(i)
    replaced=df[i].replace(-999.000,np.nan)#.mean()
    valid_mask = ~np.isnan(replaced)  # Mask to select valid (non-NaN) values
    x_valid = x[valid_mask]  # Filter x using the mask
    y_valid = replaced[valid_mask]  # Filter replaced (y) using the mask
    m, b = np.polyfit(x_valid, y_valid, 1)
    y_fit=m*x_valid+b
    all_yfit.append(y_fit)
    all_m.append(m)
    all_b.append(b)
    #replaced
    data_kept_all.append(y_valid)
    #mean.append(means)
```

### Total PM2.5 trend

```python
print(len(df['Date']))
```

```python
df.replace(-999.000, np.nan, inplace=True)
```

```python
# Group by date and sum the PM2.5 emissions
total_pm25 = df.groupby('Date')[columns].sum().reset_index()

# Calculate the total PM2.5 emissions for each date
total_pm25['Total_PM25'] = total_pm25[columns].sum(axis=1)
```

```python
total_array=(np.array(total_pm25['Total_PM25']))
```

```python
x=np.arange(0,4157,1)
```

```python
m, b = np.polyfit(x,total_array, 1)
y_fit=m*x+b
```

```python
years=np.linspace(1988,2024,4157)
```

```python
plt.plot(years,total_array,color='blue',linewidth=0.2)
yfit=plt.plot(years,y_fit,'k-',label=str(m),color='black')
plt.legend()
plt.title('Trend of total PM2.5 over 1988-2024 period',fontweight='bold',fontsize=13)
plt.xlabel('Years')
plt.ylabel('ug/m^3')
plt.ylim(-1,50)
plt.savefig('/home/areed29/summer23_notebooks/CLIM636/total_pm2.5_trend.png')
```
